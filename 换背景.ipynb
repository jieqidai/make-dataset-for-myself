```python
import os
import xml.dom.minidom
import cv2
import random

backgroungPath = "D:/SSD-1/background/"
bodyPath = "D:/SSD-1/body/"
imageResult = "D:/SSD-1/result/"
xmlResult = "D:/SSD-1/xml/"
xmlfile = "D:/SSD-1/test.xml"

backgroundImgName = os.listdir(backgroungPath)
bodyImgName = os.listdir(bodyPath)
for i in range(0,56236):
    backgroundImg = os.path.join(backgroungPath,backgroundImgName[i])
    bodyImg = os.path.join(bodyPath,bodyImgName[i])
    backgroundImage = cv2.imread(backgroundImg)
    bodyImage = cv2.imread(bodyImg)
    #print(backgroundImgName[i])
    #print(bodyImgName[i])
    
    x0 = random.randint(0, 300) 
    y0 = random.randint(0, 300)
    
    sp_body = bodyImage.shape
    h_body = sp_body[0]
    w_body = sp_body[1]
    
    sp_background = backgroundImage.shape
    h_background = sp_background[0]
    w_background = sp_background[1]
    
    roi = backgroundImage[y0:(y0+h_body),x0:(x0+w_body)]
    add_img = cv2.addWeighted(bodyImage, 0.8, roi, 0.2, 0)
    backgroundImage[y0:(y0+h_body),x0:(x0+w_body)] = add_img
    cv2.imwrite(imageResult + backgroundImgName[i],backgroundImage)
    
    dom = xml.dom.minidom.parse(xmlfile)
    root = dom.documentElement
    #获取标签对filename/width/height/xmin/ymin/xmax/ymax的值
    filename = root.getElementsByTagName('filename')
    width = root.getElementsByTagName('width')
    height = root.getElementsByTagName('height')
    xmin = root.getElementsByTagName('xmin')
    ymin = root.getElementsByTagName('ymin')
    xmax = root.getElementsByTagName('xmax')
    ymax = root.getElementsByTagName('ymax')

    #原始信息
    filename0 = filename[0]
    width0 = width[0]
    height0 = height[0]
    xmin0 = xmin[0]
    ymin0 = ymin[0]
    xmax0 = xmax[0]
    ymax0 = ymax[0]

    #修改信息
    filename0.firstChild.data = bodyImgName[i]
    width0.firstChild.data = w_background
    height0.firstChild.data = h_background
    xmin0.firstChild.data = x0
    ymin0.firstChild.data = y0
    xmax0.firstChild.data = x0+w_body
    ymax0.firstChild.data = y0+h_body

    with open(xmlResult + backgroundImgName[i][:-4] + ".xml",'w') as fh:
        dom.writexml(fh)
```
