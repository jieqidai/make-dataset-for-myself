```python
import os
import cv2 
import random 

filepath = "D:/data/nanshan/nanshanimg"
savepath = "D:/data/nanshan/nanshancrop2"

imglist =  os.listdir(filepath)
for imgname in imglist:
    imgfile = filepath + '/'+imgname
    # ,
    img=cv2.imread(imgfile)
    sp = img.shape
    h = sp[0]
    w = sp[1]
    
    size = 600 #600
    stride = 200 #
    time_h = (h - size)//stride + 1 #y
    time_w = (w - size)//stride + 1 #x
    x0 = y0 = 0 #
    
    count = 1
    for m in range(time_h+1):
        y0 = m * stride
        for n in range(time_w+1):
            x0 = n * stride
            print(x0,y0)
            cropImg = img[(y0):(y0 + size), (x0):(x0 + size)] 
            cv2.imwrite(savepath + '/' + imgname[:-4] + '_' + str(count) + '.jpg', cropImg)
            count += 1
```
